```python
import requests
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

hdr = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
       'Accept-Encoding': 'none',
       'Accept-Language': 'en-US,en;q=0.8',
       'Connection': 'keep-alive'}
       
imonumber=input("Enter IMO number: ")


URL = "https://www.vesselfinder.com/tr/vessels/details/imo"
URL =  URL.replace("imo", imonumber)
print(URL)

resp = requests.get(URL, headers = hdr)
soup = BeautifulSoup(resp.content, "html.parser")
links = soup.select('.complaint-link-for-ads')

records = []
link_soup = BeautifulSoup(resp.content, "html.parser")
```

```python
from openpyxl import Workbook, load_workbook

# create the file
wb = Workbook()
ws = wb.active
ws.append(["IMO Number", "Ship Name", "Flag", "Registered Owner", "Year of Built", "Target", "ETA", "Departure Place", "Departure Time"])
wb.save('/content/drive/MyDrive/coll/shipdata.xlsx')
```

```python
from openpyxl import Workbook, load_workbook


shipname= ""
registeredowner=""
yearofbuilt=""
target=""
estimatedtimeofarrival=""
departure_place=""
departure_time=""
flag=""
for row in link_soup.findAll('div',attrs = {'class':'vi__r1 vi__sbt'}):
  target = row.a.text
  print("target: ",target)
for row in link_soup.findAll('span',attrs = {'class':'_mcol12'}):
  estimatedtimeofarrival = row.text
  print("estimated time of arrival: ",estimatedtimeofarrival)
for row in link_soup.findAll('div',attrs = {'class':'vi__r1 vi__stp'}):
  departure_place = row.a.text
  print("departure_place: ",departure_place)
for row in link_soup.findAll('div',attrs = {'class':'_value'}):
  departure_time = row.text
  print("departure_time: ",departure_time)
for row in link_soup.findAll('div',attrs = {'class':'title-flag-icon flag-icon'}):
  flag = row['title']
  print("flag: ",flag)
for text in link_soup.select("h1.title"):
    shipname= text.get_text()
    print("Ship name: ", shipname)
for text in link_soup.select("td.v4"):
    registeredowner= text.get_text()
    print("Flag name: ",registeredowner)
for text in link_soup.select("td.v5"):
    yearofbuilt= text.get_text()
    print("year of built: ",yearofbuilt)
# re-open and append
wb = load_workbook("/content/drive/MyDrive/coll/shipdata.xlsx")
ws = wb.active
ws.append([imonumber, shipname, flag, registeredowner, yearofbuilt, target, estimatedtimeofarrival, departure_place, departure_time])
wb.save("/content/drive/MyDrive/coll/shipdata.xlsx")
```
